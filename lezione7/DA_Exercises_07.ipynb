## Exercise 07.1
The main question that arises when trying to apply data blocking is: how many steps should each block have?
The first fundamental property that has to be satisfied is the statistical independency of the block averages, namely, the mean value computed over one given block should be uncorrelated from each one of the remaining blocks average. 

First of all, after equilibrating the system, let's consider the values of energy and pressure measured after each step ($5 \times 10^5$ total steps) and let's compute the autocorrelation function. As we can see, the energy values correlation remains significant for around $\sim100/150$ steps. Hence, we obtain a general idea about correlations between values associated to single steps. 

However, the correlation that really matters is the one between the average values of each block. Therefore, we compute it and the corresponding plots seem to indicate that the correlation between adjacent blocks reaches an approximate null value between 200 and 500 block size. And this value is roughly compatible with the single step correlation value. It is worth pointing out that all our conclusion are not based on a riguorous analysis of data but simply on plots observation. Our goal is only to get a more precise idea about the blocking method and the meaning of its parameters.

Lastly, we consider how the uncertainty computed with the blocking method changes when the block size L is varied from 10 to 500. As we can see, it grows significantly at low sizes and it remains almost constant over 250/300 steps per block. Therefore, once we are sure about the absence of correlation, there is no need to further increase L, since the error remain almost constant.

The observations above are based on the plots below. Solid, liquid and gas graphs present the same charachteristics, with both pressure and energy values. The only significative difference is that while for liquid and solid the correlation data seem to differ only by a scaling factor, the same does not happen with gas, even if the overall conclusions the we can deduce are the same. This may be due to the fact that while liquids and solid have a definite volume, gases volume can vary substantially.  Thus, fixed energy and temperature fix also pressure /hence the correlation) for solid and liquid but not for gas.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

def autocorrelation(values, t_lim):
    
    tmax=values.size
    t_array=np.arange(0,t_lim)
    chi=[]
    
    sum_c=0.
    sum_d=0.

    for t_apex in range(0, tmax):
        sum_c+=values[t_apex]*values[t_apex]
        sum_d+=values[t_apex]

    denom=sum_c/tmax - (sum_d/tmax)**2


    for t in t_array:
        sum_a=0.
        sum_b1=0.
        sum_b2=0.
        for t_apex in range(0, tmax-t):
            sum_a+=values[t_apex]*values[t_apex+t]
            sum_b1+=values[t_apex]
            sum_b2+=values[t_apex+t]
        a=sum_a/(tmax-t)
        b=sum_b1*sum_b2/(tmax-t)**2
        chi.append((a-b)/denom)
    
    return chi


def error(AV,AV2,n): 
    if n==0:
        return 0
    else:
        return math.sqrt((AV2 - AV**2)/n)
    


def compute_error(L,values):
    n_blocks=int(values.size/L)
    
    x = np.arange(n_blocks)      # [0,1,2,...,N-1]
    ave = np.zeros(n_blocks)
    av2 = np.zeros(n_blocks)

    for i in range(n_blocks):
        sum = 0
        for j in range(L):
            k = j+i*L
            sum += values[k]
        ave[i] = sum/L       # block average
        av2[i] = (ave[i])**2 #  

    sum_=0
    sum2_=0
    for i in range(n_blocks):
        sum_ += ave[i] # SUM_{j=0,i} r_j
        sum2_ += av2[i] # SUM_{j=0,i} (r_j)^2
    
    av=sum_/n_blocks # Cumulative average
    av2=sum2_/n_blocks # Cumulative square average
    err = error(av,av2,n_blocks-1) # Statistical uncertainty
    
    return err
```

```python
def ex_01(state):
    #load data
    energy, pressure = np.loadtxt(rf'MonteCarlo_NVT/Ex_07.1/{state}/inst_values.dat', usecols=(0,1), delimiter=' ', unpack='true')
    N=energy.size
    n=1000

    x = np.arange(n)

    if(N>500000):
        print('error')
 

    #print first n steps
    e_cut=np.resize(energy,n)
    p_cut=np.resize(pressure,n)
    plt.plot(x,e_cut, label='energy')
    plt.plot(x,p_cut, label='pressure')
    plt.title(rf'FIRST ${n}$ (of $ 5\times 10^5$) MC steps')
    plt.grid(True)
    plt.xlabel('#steps')
    plt.legend()
    plt.show()
    
    
    
    #autocorrelation between montecarlo steps
    fig,ax=plt.subplots(1,1,figsize=(6,4))
    chi=autocorrelation(energy,200)
    t=np.arange(200)
    ax.plot(t,chi)
    ax.set_title(rf'AUTOCORRELATION BETWEEN STEPS ENERGY VALUES')
    ax.set_xlabel('#steps')
    ax.set_ylabel('autocorrelation')
    
    #autocorrelation between block averages
    array=[energy,pressure]
    title_arr=['ENERGY','PRESSURE']
    col=['blue','red']
    M=energy.size

    for index in range(2):
    
        block_size=np.array([2, 100, 200, 500, 1000])
                       
        block_vector=[]
    

        for k in block_size:
            b=M/k          #number of blocks
            block_averages=np.zeros(int(b))
        
            for i in range(int(b)):
                block_sum=0.
                for j in range(k):
                    block_sum+=array[index][i*int(k)+j]
                block_averages[i]=block_sum/k
            
            block_vector.append(block_averages)
            
        fig, axx=plt.subplots(1,block_size.size, figsize=(15,3))

        fig.suptitle(f'{title_arr[index]} AUTOCORRELATION AT DIFFERENT BLOCK SIZES', y=1.05)
        for k in range(block_size.size):
            chi=autocorrelation(block_vector[k],10)
            x=np.arange(len(chi))
            axx[k].plot(x, chi, color=col[index])
            axx[k].set_title(rf'block size = {block_size[k]}')
            axx[k].set_xlabel('#blocks')
            axx[k].set_ylabel('autocorrelation')
            
            
    #error
    L=np.linspace(10, 500, 20)#[10,100,1000,2000,5000]            # Number of throws in each block
    err=np.zeros(L.size)
    err_p=np.zeros(L.size)

    for i in range(L.size):
        err[i]=compute_error(int(L[i]),energy)
        err_p[i]=compute_error(int(L[i]),pressure)
    
    x=np.arange(err.size)

    fig, axx=plt.subplots(2,1, figsize=(10,10))

    fig.suptitle('UNCERTAINTY AS A FUNCTION OF L NUMBER OF STEP PER BLOCK', y=0.92)
    axx[0].set_xlabel('L block size')
    axx[1].set_xlabel('L block size')
    axx[0].set_ylabel('energy uncertainty')
    axx[1].set_ylabel('pressure uncertainty')
    axx[0].plot(L,err,'*')
    axx[1].plot(L,err_p,'*')
    axx[0].grid(True)
    axx[1].grid(True)
    
```

# Exercise 07.4
We now compute three quantites: potential energy, pressure and radial density distribution g(r). 
The main comment we can make is about g(r): we can observe that the solid g(r) has 4 peaks, the liquid g(r) two and the gas g(r) only one. This is due to the lattice structure typical of the solid state that dissolves by changing phase. 
The g(r) functions computed here are identical to the ones computed in exercise04. 
On the contrary the potential energy does not seem to agree to the one previously computed. The reason is that in exercise04 we did not make use of the tail corrections. Apart from this, the results are equivalent.


For the blocking method, we make use of 10 blocks of conservative size L=1000, in order to guarantee enough uncorrelation with a safe margin.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


def plot_results(state): 
    
    fig, axx=plt.subplots(2,2, figsize=(10,7))
   
    
    #potential energy
    e_con=1.65E-21
    epot_av, epot_err = np.loadtxt(rf'MonteCarlo_NVT/Ex_07.4/{state}/output.epot.0', usecols=(2,3), unpack='true')
    x=np.arange(epot_av.size)

    axx[0][0].errorbar(x,epot_av*e_con,yerr=epot_err*e_con)
    axx[0][0].grid(True)
    axx[0][0].set_title('POTENTIAL ENERGY PER PARTICLE')
    axx[0][0].set_xlabel(r'number of blocks')
    axx[0][0].set_ylabel(r'$\frac{U}{N}$  $[J]$')
    

    
    #pressure
    p_con=4.2E7
    pres_av, pres_err = np.loadtxt(rf'MonteCarlo_NVT/Ex_07.4/{state}/output.pres.0', usecols=(2,3), unpack='true')
    x=np.arange(pres_av.size)

    axx[0][1].errorbar(x,pres_av*4.2E7,yerr=pres_err*4.2E7)
    axx[0][1].grid(True)
    axx[0][1].set_title('PRESSURE')
    axx[0][1].set_xlabel(r'number of blocks')
    axx[0][1].set_ylabel(r'$p$  $[Pa]$')
    

    
    
    #g(r)
    gofr = np.loadtxt(rf'MonteCarlo_NVT/Ex_07.4/{state}/output.gave.0', skiprows=9)
    r_arr = np.loadtxt(rf'MonteCarlo_NVT/Ex_07.4/{state}/r_array.dat',usecols=(0))
    av_gofr= gofr[1::2]
    err_gofr= gofr[2::2]

    con_con=3.4E-10
    axx[1][0].remove()
    gs = axx[1, 0].get_gridspec()
    axx[1][0] = fig.add_subplot(gs[1, 0:])
    axx[1][0].errorbar(r_arr*con_con,av_gofr,yerr=err_gofr)
    axx[1][0].set_title("$g(r)$")
    axx[1][0].set_xlabel("$r (m)$")
    axx[1][0].set_ylabel("$g(r)$")
    axx[1][0].grid(True)
    axx[1][1].remove()
    
    fig.tight_layout()
    
```

### solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)

```python
plot_results('solid')  
```
