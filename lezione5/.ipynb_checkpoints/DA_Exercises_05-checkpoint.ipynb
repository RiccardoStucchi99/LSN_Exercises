## Exercise 05.1

The exercise asks to compute the average radius associated to two different quantum states of the Hydrogen atom by exploiting the Metropolis sampling method.

Before computing the requested values, let's properly set the sampling algorythm. 
First of all, in order to follow the 50% empirical rule, we vary the step of the uniform transition probability. In particular, a 1.0 step and a 2.0 step are chosen for the sampling of $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$, respectively. 
Then, by observing how the moduli r of the sampled points are distributed, we can deduce how to choose the position of the initial point. In fact, the moduli tend to be confined into a limited range corresponding to the most probable values of the target distribution. For instance, in the $|\Psi_{1,0,0}(x,y,z)|^2$ case, this range is approximately [0,4]. If we choose a point far from that range, the algorythm after some equilibration steps returns to the "equilibrium" range but the sampled values during the equilibration should be discarded. Therefore it is better to chose an initial point inside the equilibrium range so that all points sampled can be used for the averages computation.

If we make a histogram (Figures 2 & 3) of the moduli of the sampled points and we observe it after a different number of steps, we can understand how Metropolis operates. In particular, the asympotical evolution of the histogram towards the target distribution (here the target distribution is the radial probability) is showed.  After about 10000 steps the sampling can be considered satisfactory but more than 100000 are needed for a "perfect" sampling. Because the best estimation of our computation is given by $\frac{1}{N}\sum_{i=0}^{N}{O_i}$, with $N$ the total number of steps and $O_i$ the single observable evaluations, we should choose N large enough so that the resulting Metropolis distribution is as similar as possible to the target one. Let's consider for example figure 2 with N=5000. In this case we can see some bulges when $r>r_{peak}$ and some recesses when $r<r_{peak}$. Hence we would have an over-estimation of $\langle r \rangle$ since the generated distribution has slightly larger $r$-concentration at higher $r$. Therefore we will use a conservative value of $N=1000000$, 10 times larger than the value $N=100000$ where a satisfactory sampling is first observed.

As regards the size of the blocks, the main condition to be satisfied is the block averages uncorrelation. We will go into details about this in a future exercise. However, since for a "graphical" reason we would like to have 100 blocks, the given $N$ and 100 blocks implies block size $L=10000$. It should be enough to guarantee the mentioned uncorrelation.

Moreover, as already mentioned we initially make use of a uniform transition probability. Then, we also introduce a gaussian transition probability (for simplicity $\sigma=unif.step$). What we observe at first is that the acceptance rate decreases($\sim0.6\rightarrow\sim0.4$). This is not surprising. In fact, compared to a uniform distribution, a gaussian one proposes more distant points where the target p(x) is in general smaller. 

Once that the Metropolis algorythm is ready, the quantities $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ are computed. With uniform transition probability, 1000000 points are sampled and divided into 100 blocks of size 10000. The results agree with the expected values (Figure 4).
Then, the transition probability is changed with a gaussian ($\sigma=unif.step$). Without modifying the number and the size of the blocks, the results are less precise than before. This is due to the smaller acceptance rate and the consequent reduced efficiency. Thus, to obtain better results we increase the total number of points by a factor of 10 (Figure 5: the size of the blocks is now 100000).

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

from mpl_toolkits.mplot3d.axes3d import get_test_data

x_100, y_100, z_100, r_100, x_210, y_210, z_210, r_210 = np.loadtxt('ex_05.1/ex_05.1.block_test.dat', delimiter=' ', unpack='true')
n=x_100.size
```

```python
fig = plt.figure(figsize=(12,6))

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter(x_100, y_100, z_100,c=z_100, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)


ax = fig.add_subplot(1, 2, 2, projection='3d')
ax.scatter(x_210, y_210, z_210, c=z_210, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
fig.suptitle('Figure 1: SAMPLED POINTS, $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$')
```

```python
def f_100(r):
    return 4*r**2*np.exp(-2*r)

def f_210(r):
    return r**4*np.exp(-r)/24.

ep=[6,15]
```

```python
def plot_distr(arr, n, f, idx):
    
    if(n<=r_100.size):
        a=np.arange(n)
        arr=np.resize(arr,n)
        
    
    fig = plt.figure(figsize=(10,6))
    fig.suptitle(f'Figure {idx+2}')
    ax = fig.add_subplot(projection='3d')
    nbins = 100
    ys = arr


    hist, bins = np.histogram(ys, bins=nbins, density='True')
    xs = (bins[:-1] + bins[1:])/2
    dx=bins[1]-bins[0]

    obj1=ax.bar(xs, hist, width=dx, zdir='y', alpha=0.8)
    obj1.set_label('computed radial distribution')

    ax.set_xlabel('r value')
    ax.set_ylabel('# step')
    ax.set_zlabel('occurences')
    ax.invert_xaxis()
    ax.view_init(30, 40)

    yr=np.arange(n)
    zr=np.zeros(n)
    ax.plot(arr, yr, zr)
    
    r=np.linspace(0,ep[idx], 200)
    yr=np.ones(r.size)*0
    obj2,=ax.plot(r, yr, f(r),'r')
    obj2.set_label('expected radial distribution')
    
    ax.legend(loc=1)

    plt.show()
```

### Moduli histogram $|\Psi_{1,0,0}(x,y,z)|^2$

```python
plot_distr(r_100, 1000000, f_100, 0)
```

### Moduli histogram $|\Psi_{2,1,0}(x,y,z)|^2$

```python
plot_distr(r_210, 10000, f_210, 1)
```

### $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$ computation

```python
prog_av_100, error_100 = np.loadtxt('ex_05.1/ex_05.1.psi_100.dat', usecols=(0,1), delimiter=' ', unpack='true')
prog_av_210, error_210 = np.loadtxt('ex_05.1/ex_05.1.psi_210.dat', usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(prog_av_100.size)

fig, ax=plt.subplots(1,2, figsize=(15,5))
ax[0].errorbar(x,prog_av_100,yerr=error_100)
ax[1].errorbar(x,prog_av_210,yerr=error_210)

ax[0].set_xlabel('number of blocks')
ax[0].set_title(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
ax[0].set_ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$ (Bohr units)')
ax[1].set_xlabel('number of blocks')
ax[1].set_ylabel(r'$\left\langle r \right\rangle_{\Psi_{2,1,0}}$ (Bohr units)')
ax[1].set_title(r'$\left\langle r \right\rangle_{\Psi_{2,1,0}}$')
ax[0].grid()
ax[1].grid()

fig.suptitle('Figure 4: RADIUS BLOCK AVERAGE AND UNCERTAINTY')
print('final value and uncertainty')
print(f'psi(1,0,0): {prog_av_100[-1]} +- {error_100[-1]}')
print(f'psi(2,1,0): {prog_av_210[-1]} +- {error_210[-1]}')
```

```python
prog_av_100, error_100 = np.loadtxt('ex_05.1/ex_05.1.psi_100_G.dat', usecols=(0,1), delimiter=' ', unpack='true')
prog_av_210, error_210 = np.loadtxt('ex_05.1/ex_05.1.psi_210_G.dat', usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(prog_av_100.size)

fig, ax=plt.subplots(1,2, figsize=(15,5))
ax[0].errorbar(x,prog_av_100,yerr=error_100)
ax[1].errorbar(x,prog_av_210,yerr=error_210)

ax[0].set_xlabel('number of blocks')
ax[0].set_title(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$')
ax[0].set_ylabel(r'$\left\langle r \right\rangle_{\Psi_{1,0,0}}$ (Bohr units)')
ax[1].set_xlabel('number of blocks')
ax[1].set_ylabel(r'$\left\langle r \right\rangle_{\Psi_{2,1,0}}$ (Bohr units)')
ax[1].set_title(r'$\left\langle r \right\rangle_{\Psi_{2,1,0}}$')
ax[0].grid()
ax[1].grid()

fig.suptitle('Figure 5: RADIUS BLOCK AVERAGE AND UNCERTAINTY, Gaussian transition probability')
print('final value and uncertainty')
print(f'psi(1,0,0): {prog_av_100[99]} +- {error_100[99]}')
print(f'psi(2,1,0): {prog_av_210[99]} +- {error_210[99]}')
```
