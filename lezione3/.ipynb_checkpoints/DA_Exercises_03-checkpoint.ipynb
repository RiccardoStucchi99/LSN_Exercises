## Exercise 03.1
The following exercise consists in a simple application of an econophysics model that allows to estimate, under certain hypotheses, the price of Plain Vanilla - European options. This kind of option is the most basic one and it gives the buyer the right to sell or purchase an underlying asset for a prescribed price K at the expiry date T.

The Black-Scholes theory, by assuming market idealities and a $GBM(r,\sigma^2)$ time evolution of the asset price $S(t)$ ($r$ is the constant risk-free interest rate, $\sigma$ is the volatility), obtains analytical solutions for the price of the options. At time $t<T$:
- Call-option price:
$$ C\left[ S(t),t \right] = S(t) N(d_1) -K e^{-r(T-t)} N(d_2) $$
- Put-option price:
$$ P\left[ S(t),t \right] = S(t) \left[ N(d_1)-1\right] -K e^{-r(T-t)} \left[ N(d_2)-1\right]$$

where 
$$d_1 = \frac{1}{\sigma\sqrt{T-t}}\left[ \ln \left( \frac{S(t)}{K} \right) + \left( r+\frac{\sigma^2}{2} (T-t) \right)\right]$$

where $d_2 = d_1 - \sigma\sqrt{T-t}$, and where
$$ N(x) = \int_{-\infty}^{x} dz \, e^{-\frac{z^2}{2}} = \frac{1}{2} \left[ 1+ {\rm erf} \left( \frac{x}{\sqrt{2}}\right) \right]$$
is the cumulative of Gaussian $\sim \mathcal{N}(\mu=0,\sigma^2=1)$.



Unfortunately, under less ideal assumption, analytic solutions are no longer available and another price evaluation method is required. This is where Monte-Carlo methods turn out to be useful. In fact, a fair option pricing at time $t=0$ is simply the expected profit at expiry (discounted by a Bank deposit risk free factor $exp(-rT)$), which corresponds to the following average values:
- Call-option price:
$$ C\left[ S(t),t \right] = \left\langle e^{-rT} max[0,S(T) - K] \right\rangle $$
- Put-option price:
$$ P\left[ S(t),t \right] = \left\langle e^{-rT} max[0, K - S(T)] \right\rangle $$

Hence, by Monte-Carlo simulating the $S(t)$ stochastic evolution, prices can be easily computed.



In order to compare analytical and Monte-Carlo predictions,  we assume, just like in the Black-Scholes theory, that the asset price at time t is given by a geometric brownian motion with drift $r$ and volatility $\sigma$ ($GBM(r,\sigma^2)$). The asset price at t=T can be sampled by means of two equivalent methods. 
The first one consists in directly sampling S(T) as:
$$
S(T) = S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma \mathcal{N}(0,1) \sqrt{T}\right]$$
where $\mathcal{N}(0,1)$ is a Gaussian distribution with $\mu=0$ and $\sigma=1$. 

On the other hand, S(T) could be discretized into $n$ sub-intervals $t_0=0<t_1 < t_2 < â€¦ < t_n=T$ so that:
$$S(t_{i+1}) = S(t_i) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}\right]$$
 
As already stated, the two methods are equivalent:
$$
\begin{align}
S(T) &= \prod_{i=0}^{n-1}{S(t_i)} \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{n}-t_{n-1})+\sigma \mathcal{N}(0,1)\sqrt{t_{n}-t_{n-1}}\right]\\
&= S(0) \prod_{i=0}^{n-1} \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}\right]\\
&= S(0) \exp\left\{\sum_{i=0}^{n-1} {\left[ \left(\mu-\frac{1}{2}\sigma^2\right)(t_{i+1}-t_i)
    +\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}\right]}\right\}\\
&= S(0) \exp\left[ \left(\mu-\frac{1}{2}\sigma^2\right)T+\sigma \mathcal{N}(0,1) \sqrt{T}\right]
\end{align}
$$   

Where for the last equality it was used: $\sum_{i=0}^{n-1}{\sigma \mathcal{N}(0,1)\sqrt{t_{i+1}-t_i}} = \sigma \mathcal{N}(0,\sigma^2=n)\sqrt{t_{i+1}-t_i} = \sigma \sqrt(n) \mathcal{N}(0,1) \sqrt{t_{i+1}-t_i} = \sigma \mathcal{N}(0,1) \sqrt{T}$.


#### Comments on the results:
Given the arguments above, 4 quantities (progressive average + uncertainty) are computed by sampling the GBM: European call-option prices, $C[S(0),0]$ (direct - Figure 1 ; discretized - Figure 3), and put-option prices, $P[S(0),0]$ (direct - Figure 2 ; discretized - Figure 4).
The following option parameters are chosen:
- asset price at $t=0$: $S(0)=100$
- delivery time: $T=1$
- strike price: $K=100$
- risk-free interest rate: $r=0.1$
- volatility: $\sigma=0.25$
- number of intervals for GBM discretization: n=100

In addition, for each quantity N=100000 Monte-Carlo throws are made and divided into M=1000 blocks in order to apply the blocking method.
By subtracting the analytical values (call:  14.975790778311286, put:  5.4595325819072364) from the estimated one, the agreement between Black-Scholes theory and Monte-Carlo simulation is shown.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np


prog_av, error = np.loadtxt('ex03.1/call.dat', usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(prog_av.size)

plt.errorbar(x,prog_av-14.975790778311286,yerr=error)
plt.title(r"Figure 1: Estimation of the price of a call option, $C[S(0),0]$")
plt.xlabel('#blocks')
plt.ylabel(r'$C[S(0),0]- (C[S(0),0])_{analytical}$')
plt.grid(True)
plt.show()


prog_av, error = np.loadtxt('ex03.1/put.dat', usecols=(0,1), delimiter=' ', unpack='true')
x = np.arange(prog_av.size)

plt.errorbar(x,prog_av-5.4595325819072364,yerr=error)
plt.title(r"Figure 2: Estimation of the price of a put option, $P[S(0),0]$")
plt.xlabel('#blocks')
plt.ylabel(r'$P[S(0),0]- (P[S(0),0])_{analytical}$')
plt.grid(True)
plt.show()
```
