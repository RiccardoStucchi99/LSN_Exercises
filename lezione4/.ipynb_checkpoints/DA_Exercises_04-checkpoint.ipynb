# Exercises 04
A system of n=108 particles/molecules interacting through a short range Lennard Jones potential
$$
V(r)=4\epsilon\left[ \left(\frac{\sigma}{r}\right)^{12} -\left(\frac{\sigma}{r}\right)^{6} \right]
$$

is simulated by means of a Molecular Dynamics code. After applying a few optimization to the provided code, simulations are run and some quantities of interest such as temperature and internal energy are computed in different configurations of the system.
In the exercise the n=108 particles are chosen to be Argon atoms ($\sigma = 0.34$ nm, $\epsilon/k_B = 120$ K, $m=39.948$ amu).

## Exercise 04.1

After implementing the restart option, the first step consists in equilibrating the system at different phases and temperatures. The equilibration process is the same for each of the three phases:

- the system is initialized through the Input() function, which given a set of starting positions (a fcc lattice configuration is used) executes the following tasks: -set the system parameters; - randomly extracts molecules velocity at $t=0$; - corrects velocities so that the center of mass velocity is null -scales velocities to match a given temperature.

- the system is then simulated for a number of steps n by means of the Move() function based on the Verlet algorythm

- the system is restarted through Restart_Input that does the same operations of the Input(), apart from the initial velocities calculation that is now made by using the final configuration of the previous step simulation.

- Step 2 and 3 are repeated until equilibrium is reached.


The parameters chosen for each phase are:

|            | Temp | rho  | r_cut | target Temp |
|------------|------|------|-------|-------------|
| **solid**  | 1.4  | 1.1  | 2.2   |0.8          |
| **liquid** | 1.9  | 0.8  | 2.5   |1.1          |
| **gas**    | 1.1  | 0.05 | 5.0   |1.2          |


It is worth noting the difference between 'Temp' and 'target Temp' given in the table above. The former is only used in the Input function and its value is given so that the system reaches the target temperature, which is the desired equilibrium tempearature value, already after the first equilibration step. In other words, if temp were different, it would take a few more iterations of the steps described above to reach equilibrium. 

Regarding the specific number of steps needed for equilibrium, for all the phases an Input start plus a Restart, both with n=1000 steps, is enough to reach and check equilibration.

Lastly, to verify that a given set of parameters corresponds to a certain state of matter the positions of the particles are printed at different time steps and visualized with Ovito. The different behaviours that are observed agree with the expected phase. 

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

c=[0.8,1.1,1.2]
state=["solid","liquid","gas"]
fig, axx=plt.subplots(3,1, figsize=(6,10))

   
for i in range(3):
    temp= np.loadtxt(f'MolecularDynamics_NVE/{state[i]}/output_temp.dat', usecols=(0), delimiter=' ', unpack='true')
    x = np.arange(temp.size)
    y = np.ones(temp.size)*c[i]
    axx[i].plot(x,temp)
    axx[i].plot(x, y)
    axx[i].grid(True)
    axx[i].set_xlabel('Measure (1 measure every 10 steps)')
    axx[i].set_ylabel(r'Temperature ($\frac{\epsilon}{k_{B}}$)')
    axx[i].set_title(f"Figure {i+1}: "+state[i]+" equilibration")
    axx[i].legend(['computed values', 'target value'])
    
plt.tight_layout()

#1000 solid 2000 liquid 4000(500) gas
```

## Exercise 04.3
Once the system is equilibrated, average energy (kinetic, potential and total) and average temperature can be computed through the usual blocking method. Specifically, 10 blocks of 100 measures (10 steps per measure) are used.

As requested, the values computed in LJ units are converted into SI units. Here is the conversion table:

| Physical quantity         &nbsp;&nbsp;&nbsp;&nbsp;          |  &nbsp;&nbsp;&nbsp;   Unit  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;   | &nbsp;&nbsp;&nbsp;   Value for Ar  &nbsp;&nbsp;&nbsp;  &nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;  |
|-------------------------------------|--------------|-----------------------|
| length                              | $σ$          | $3.4×10^{−10}m$   |
| energy                              | $ε$          | $1.65×10^{−21}J$   |
| mass                                | $m$          | $6.69×10^{−26}kg$  |
| time                                | $σ(m/ε)^{1/2}$ | $2.17×10^{−12}s$  |
| velocity                            | $(ε/m)^{1/2}$  | $1.57×10^{2}m/s$   |
| force                               | $ε/σ$        | $4.85×10^{−12}N$   |
| pressure                            | $ε/σ^3$    | $4.20×10^{7}N·m^{−2}$ |
| temperature                         | $ε/k_B$       | $120K$            |
 
There is an extra plot representing the function g(r). It will be useful in the 7th lecture.

```python
epot = {
  "obs_name" : 'epot',
  "obs_name_ext" : 'POTENTIAL ENERGY PER PARTICLE',
  "conversion" : 1.65E-21,
  "x_title" : 'number of blocks',
  "y_title" : r"$\frac{U}{N}$   $(J)$"
}


ekin = {
  "obs_name" : 'ekin',
  "obs_name_ext" : 'KINETIC ENERGY PER PARTICLE',
  "conversion" : 1.65E-21,
  "x_title" : 'number of blocks',
  "y_title" : r"$\frac{K}{N}$  $(J)$"
}


etot = {
  "obs_name" : 'etot',
  "obs_name_ext" : 'TOTAL ENERGY PER PARTICLE',
  "conversion" : 1.65E-21,
  "x_title" : 'number of blocks',
  "y_title" : r"$\frac{E}{N}$  $(J)$"
}

temp = {
  "obs_name" : 'temp',
  "obs_name_ext" : 'TEMPERATURE',
  "conversion" : 120,
  "x_title" : 'number of blocks',
  "y_title" : r"$T$ $(K)$"
}
```

```python
def plot_results(state):
    
    obs=[epot,ekin,etot,temp]
    
    fig, axx=plt.subplots(3,2, figsize=(10,10))
   
    for i in range(4):
        obs_av, obs_err = np.loadtxt(rf'MolecularDynamics_NVE/{state}/ave_{obs[i]["obs_name"]}.dat', usecols=(0,1), delimiter=' ', unpack='true')
        x = np.arange(obs_av.size)
        i2=i%2
        i1=(i//2)%2
        axx[i1][i2].errorbar(x, obs_av*obs[i]["conversion"], obs_err*obs[i]["conversion"])
        axx[i1][i2].set_title(obs[i]["obs_name_ext"])
        axx[i1][i2].set_xlabel('number of blocks')
        axx[i1][i2].set_ylabel(obs[i]["y_title"])
        axx[i1][i2].grid(True)
        
    
    
    gofr = np.loadtxt(rf'MolecularDynamics_NVE/{state}/gave.dat', skiprows=9)
    r_arr = np.loadtxt(rf'MolecularDynamics_NVE/{state}/r_array.dat',usecols=(0))
    av_gofr= gofr[1::2]
    err_gofr= gofr[2::2]

    con_con=3.4E-10
    axx[2][0].remove()
    gs = axx[2, 0].get_gridspec()
    axx[2][0] = fig.add_subplot(gs[2, 0:])
    axx[2][0].errorbar(r_arr*con_con,av_gofr,yerr=err_gofr)
    axx[2][0].set_title("$g(r)$")
    axx[2][0].set_xlabel("$r (m)$")
    axx[2][0].set_ylabel("$g(r)$")
    axx[2][0].grid(True)
    axx[2][1].remove()
    
    fig.tight_layout()
```

### solid phase: $\rho^\star = 1.1$, $T^\star = 0.8$ (cut-off radius: $r_c = 2.2$)

```python
plot_results("solid")
```

### liquid phase: $\rho^\star = 0.8$, $T^\star = 1.1$ (cut-off radius: $r_c = 2.5$)

```python
plot_results("liquid")
```

### gas phase: $\rho^\star = 0.05$, $T^\star = 1.2$ (cut-off radius: $r_c = 5.0$)

```python
plot_results("gas")
```
